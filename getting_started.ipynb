***data Cleaning***

```python
data.isnull().sum()

# drop missing value

data.dropna(inplace=True)
#data.isnull().sum()
# we have removed missing values
data.duplicated().sum() 
# we haave 12 duplicate values 
#data[data.duplicated()]
data.drop_duplicates(inplace=True)
data.duplicated().sum() 


```

***Now we are checking datatypes of columns***

```python
# we changing data types according to our use
data.dtypes
data['id']=data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object)
data.dtypes
```
